```python
def get_and_save_cuo():
    
    for ind___,file in enumerate(filename_data.keys()):
    
        


        index_ = ['БЕ', 'Целевая катег-я запа', 'Завод потребности',
               'Завод потребности_2', 'Расч. завод пополн.', 'Группа материало',
               'Подгруппа материалов', 'Подгруппа материала 2', 'Материал',
               'Материал2', 'Материал расчетный', 'Материал расчетный2', 'Модуль',
               'Цена плановая', 'Комплекс', 'Комплекс для ЦУО', 'БЕИ']


        index_key = ['БЕ', 'Целевая катег-я запа', 'Завод потребности',
               'Завод потребности_2', 'Расч. завод пополн.', 'Группа материало',
               'Подгруппа материалов', 'Подгруппа материала 2', 'Материал',
               'Материал2', 'Материал расчетный', 'Материал расчетный2', 'Модуль',
               'Цена плановая', 'Комплекс', 'Комплекс для ЦУО', 'БЕИ','Дата','Период']

        mas = r'\d\d.\d\d.\d\d\d\d'
        
        

        print('Начинаем загрузку файла ' +file)

        df =  pd.read_excel(directory+file+'.xlsx',sheet_name='Group')
        
#         df = df[:10000].copy()
        
        for col__ in index_:
            df[col__] = df[col__].astype('str')
            

        print('Загрузку файла завершили, начинаем сбор файла')


        for col in index_: df[col].fillna('',inplace = True)
            
        s = list(set([re.findall(mas,x)[0] for x in df.columns if len(re.findall(mas,x))>0]))
        it = 0
        for val_ in s:
            col = [x for x in df.columns if x.find(val_)!=-1]
            index_.extend(col)

            new_df = df[index_]

            new_df['Дата'] = val_
            new_df['Период'] = filename_data[file]

            for col___ in col:
                new_df.rename(columns={col___: col___.replace(re.findall(mas,col___)[0],'').replace(' ', '')},inplace=True)

            if it==0:

                df_all = new_df

            else:

                df_all = pd.merge(left=df_all,right=new_df,\
                 on=index_key,how='outer')


            index_ = index_[:len(index_)-len(col)].copy()
#             print(val_)
#             print(len(df_all))
            it = it+1
            
        for col__ in index_:
            df_all[col__] = df_all[col__].astype('str')
            
        print('Сбор файла' + file + 'закончили, начинаем разметку столбцов')
        
        del df        
        vn = [ x for x in df_all if x not in index_key]

        mask = []
        map_ = {}
        for ind_, col in enumerate(vn):
            find_mask  =re.findall(r'_\D',col)
            if len(find_mask)>0:
                mask.append(col.replace(find_mask[0],''))
            else:
                mask.append(col)

        for col in list(set(mask)):
            map_[col] = list(set([x for x in vn if x.find(col)==0 ]))
        #     print([x for x in vn if x.find(col)==0 and len(re.findall(r'_\D',x))!=0   ])


        for map_key in map_.keys():
            if len(map_[map_key])>1:
                df_all[map_key+'__'] =   df_all[ map_[map_key] ].sum(axis = 1)
                df_all.drop(map_[map_key],inplace = True,axis = 1)

#         if ind___ != len(filename_data.keys())-1:
#             ch_l = [x for x in df_all.columns if x.find('__')!=-1 and x.find(r',') ]
# #             ls = [x[:x.find(r',')]+' '+filename_data[file]+' '+x[x.find(r','):] for x in df_all.columns if x.find('__')!=-1 and x.find(r',') ]
#             cols_pr = [x.replace('__','') for x in ch_l]
#             df_all.rename(columns={ch_l[ind]: cols_pr[ind] for ind,x in enumerate(ch_l)},inplace=True)
        
        for i in index_:
                df_all[i] = df_all[i].apply(lambda x: str(x).replace('.0','')).astype('str')

        df_all['Цена плановая'] = df_all['Цена плановая'].replace('',0)
        df_all['Цена плановая'] = df_all['Цена плановая'].astype('float64')

        print('Разметку закончили, сохраняем файл')
        df_all.to_csv(directory+str(ind___)+'.csv',encoding = 'utf-8-sig',index = False,decimal=',',float_format='%.15f')
        del df_all
```
